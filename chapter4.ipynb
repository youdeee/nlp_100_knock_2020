```python
import MeCab

t = MeCab.Tagger()
with open("neko.txt") as f:
    text = f.read()
with open("results/neko.txt.mecab", mode='w') as f:
    f.write(t.parse(text))
```

```python
# 30

tokens = []
with open("results/neko.txt.mecab") as f:
    token = f.readline()
    while token != "":
        if len(token.split("\t")) > 1:
            tokens.append({
                "surface": token.split("\t")[0],
                "base": token.split("\t")[1].split(",")[6],
                "pos": token.split("\t")[1].split(",")[0],
                "pos1": token.split("\t")[1].split(",")[1]
            })
        token = f.readline()
```

```python
# 31

list(set([token["surface"] for token in tokens if token["pos"] == "動詞"]))
```

```python
# 32

list(set([token["base"] for token in tokens if token["pos"] == "動詞"]))
```

```python
# 33

list(set([tokens[i]["surface"] + tokens[i+1]["surface"] + tokens[i+2]["surface"] for i in range(len(tokens) - 2) if tokens[i]["pos"] == "名詞" and tokens[i+1]["surface"] == "の" and tokens[i+2]["pos"] == "名詞"]))
```

```python
# 34

prev = []
result = []
for i in range(len(tokens)):
    if tokens[i]["pos"] == "名詞":
        prev.append(tokens[i]["surface"])
    else:       
        if len(prev) > 1:
            result.append("".join(prev))
        prev = []
list(set(result))
```

```python
# 35

from itertools import groupby

sorted_tokens = sorted(tokens, key=lambda m: m['surface'], reverse=True)
counts = [(key, len(list(group))) for key, group in groupby(sorted_tokens, key=lambda m: m['surface'])]
counts = sorted(counts, key=lambda m: m[1], reverse=True)
```
