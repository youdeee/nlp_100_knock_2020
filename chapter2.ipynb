```python
# 13

col1 = pd.read_csv("col1.txt", header=None)
col2 = pd.read_csv("col2.txt", header=None)

# concat([], axis=1) でもいけるっぽい
pd.merge(col1, col2, left_index=True, right_index=True).to_csv("popular-names_name_gender.txt", sep="\t", header=None, index=False)
```

```python
# 15

N = 10
df[:-(N+1):-1]#.to_csv("popular-names_tail.txt", sep="\t", header=None, index=False)
```

```python
# 16

N = 10
n_for_line = -(-len(df) // N)
for i in range(N):
    k = i * n_for_line
    df[k:k+n_for_line].to_csv(f"split-{i}.txt", sep="\t", header=None, index=False)
```

```python
# 19

res = pd.DataFrame()
for name in list(df[0].unique()):
    data = [name, df[0].loc[df[0] == name].count()]
    res = pd.concat([res, pd.DataFrame([data])], ignore_index=True)
res.sort_values(1, ascending=False)
```
